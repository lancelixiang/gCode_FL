```python
import torch
import torchvision.models as models
```

```python
# 获取models下的所有属性和方法，并过滤掉非模型类
model_names = [name for name, obj in vars(models).items() if isinstance(obj, type) and issubclass(obj, torch.nn.Module) and obj.__module__.startswith('torchvision.models')]

# 打印所有模型名称
print(model_names)
```

```python
# 分类模型
classification_models = {name: obj for name, obj in vars(models).items() if 'resnet' in name.lower() or 'vgg' in name.lower() or 'alexnet' in name.lower() or 'inception' in name.lower() or 'densenet' in name.lower() or 'mobilenet' in name.lower() or 'vit' in name.lower()}
print(classification_models.keys())
```

```python
# 检测模型
detection_models = {name: obj for name, obj in vars(models.detection).items() if 'faster' in name.lower() or 'mask' in name.lower() or 'retina' in name.lower()}
print(detection_models.keys())
```

```python
# 分割模型
segmentation_models = {name: obj for name, obj in vars(models.segmentation).items()}
print(segmentation_models.keys())
```
